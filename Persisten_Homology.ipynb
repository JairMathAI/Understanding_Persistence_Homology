<h1><b>K-simplicies</h1>
</center><br><br>
    
<center>
<h2>Is a convex hull of $k + 1$ vertices. Suppose $k + 1$ points ${\displaystyle u_{0},\dots ,u_{k}}$ are affinely independ, ie, ${\displaystyle u_{1}-u_{0},\dots ,u_{k}-u_{0}}$ are linearly independent. </h2>
<br><br>
<h2>Then, the simplex determined by them is the set of points</h2>
</center><br><br> 

<center>
<h2>${\displaystyle C=\left\{\theta _{0}u_{0}+\dots +\theta _{k}u_{k}~{\Bigg |}~\sum _{i=0}^{k}\theta _{i}=1{\mbox{ and }}\theta _{i}\geq 0{\mbox{ for }}i=0,\dots ,k\right\}.}$</h2>
</center><br><br> 

<center>
<h1><b>Simplicial complex</h1>
</center><br><br>

<center>
<h2>A simplicial complex ${\displaystyle {\mathcal {K}}}$ is a set of simplices that satisfies the following conditions:</h2>
<br><br>

<h2>Every face of a simplex from  ${\displaystyle {\mathcal {K}}}$ is also in ${\displaystyle {\mathcal {K}}}$.</h2>
<br><br>

<h2>The non-empty intersection of any two simplices ${\displaystyle \sigma _{1},\sigma _{2}\in {\mathcal {K}}}$ is a face of both ${\displaystyle \sigma _{1}}$ and ${\displaystyle \sigma _{2}}$.</h2>
<br><br>
</center>
    
<center>
<img src="source/img/5.PNG" width="1100" height="600"/>
</center>

<center>

<center>
<h1><b>Chain Complexes</h1>
</center><br><br>

<center>
<h2>Sequence $( \mathscr{C},\partial )$ of abelian groups $\mathscr{C}_{n}$ (chains) and group homomorphisms $\partial_{n}$ (boundary maps) such that: $\partial_{n}\circ\partial_{n+1}=0$</h2>
</center> <br>

<center>
<h2>${\cdots \longrightarrow \mathscr{C}_{n+1}{\stackrel {\partial_{n+1}}{\longrightarrow }}\mathscr{C}_{n}{\stackrel {\partial_{n}}{\longrightarrow }}\mathscr{C}_{n-1}{\stackrel {\partial_{n-1}}{\longrightarrow }}}\cdots$</h2>
</center>

<center>
<h1><b>Homology (Finaly!!! 😂)</h1>
</center><br><br>

<center>
<h1><b>${\displaystyle n}-$th group of cycles $Z_{n}$</h1>
</center><br>

<center>
<h2>$\displaystyle Z_{n} := \ker \partial_{n}:=\{c\in \mathscr{C}_{n}\,|\;\partial_{n}(c)=0\}$</h2>
</center><br>

<center>
<h1><b>${\displaystyle n}-$th group of boundaries ${\displaystyle B_{n}}$</h1>
</center><br>
 
<center>
<h2>${\displaystyle B_{n}:=\mathrm {im} \,\partial_{n+1}:=\{\partial_{n+1}(c)\,|\;c\in \mathscr{C}_{n+1}\}}$</h2>
</center><br>

<center>
<h1><b>${\displaystyle n}-$th homology group ${\displaystyle H_{n}}$</h1>
</center><br>

<center>
<h2>${\displaystyle H_{n}=Z_{n}/B_{n}} = \ker \partial_{n} / \mathrm {im} \,\partial_{n+1}$</h2>
</center> <br>

<center>
<img src="source/img/7.png" width="1100" height="600"/>
</center>

<center>
<h1>${\mathbf{0} {\stackrel {\partial_{3}}{\longrightarrow }} \mathbb{Z}_2 {\stackrel {\partial_{2}}{\longrightarrow }}\mathbb{Z}_2^3{\stackrel {\partial_{1}}{\longrightarrow }}\mathbb{Z}_2^3{\stackrel {\partial_{0}}{\longrightarrow }}}0$</h1>
</center>

$\hspace{8mm}$
<center>
<h1>$\partial_0(a)=\partial_0(b)=\partial_0(c)=0$</h1>
</center>

$\hspace{8mm}$
<center>
<h1>$\partial_1(v_0)=a+b$</h1>
</center>
<center>
<h1>$\partial_1(v_1)=b+c$</h1>
</center>
<center>
<h1>$\partial_1(v_2)=c+a$</h1>
</center>

$\hspace{8mm}$
<center>
<h1>$\partial_2(f)=v_0+v_1+v_2$</h1>
</center>

$\hspace{8mm}$
<center>
<h1>$\partial_3(0)=0$</h1>
</center>

$\hspace{8mm}$
<center>
<h1>${\displaystyle H_{2}(\mathbb{D})=Z_{2}/B_{2}}=\ker \partial_{2}/\mathrm {im} \,\partial_{3} = \ker \partial_{2}/\{0\} = \{0\}$</h1>
</center>

$\hspace{8mm}$
<center>
<h1>${\displaystyle H_{1}(\mathbb{D})=Z_{1}/B_{1}}=\ker \partial_{1}/\mathrm {im} \,\partial_{2}$</h1>
</center>

$\hspace{8mm}$
<center>
<h1>$\displaystyle \ker \partial_{1}$</h1>
</center>

$\hspace{8mm}$
<center>
<h1>$\displaystyle \partial_1(c)=\alpha(a+b)+\beta(b+c)+\gamma(c+a)=0$</h1>
</center>

$\hspace{8mm}$
<center>
<h1>$\displaystyle \partial_1(c)=(\alpha+\gamma)a+(\alpha+\beta)b+(\beta+\gamma)c=0$</h1>
</center>

$\hspace{8mm}$
<center>
<h1>$\displaystyle \alpha=\beta=\gamma \because 0 \text{ or } 1$</h1>
</center>

$\hspace{8mm}$
<center>
<h1>$\displaystyle c=\alpha(v_0+v_1+v_2) \Rightarrow \text{spawn} (v_0+v_1+v_2)$</h1>
</center>

$\hspace{8mm}$
<center>
<h1>$\displaystyle \mathrm {im} \,\partial_{2} \Rightarrow \text{spawn} (v_0+v_1+v_2)$</h1>
</center>

$\hspace{8mm}$
<center>
<h1>$\displaystyle H_{1}(\mathbb{D})= \text{spawn} (v_0+v_1+v_2)/\text{spawn} (v_0+v_1+v_2) = \{\text{spawn} (v_0+v_1+v_2)\} \cong \{0\}$</h1>
</center>

$\hspace{8mm}$
<center>
<h1>${\displaystyle H_{0}(\mathbb{D})=Z_{0}/B_{0}}=\ker \partial_{0}/\mathrm {im} \,\partial_{1}$</h1>
</center>

$\hspace{8mm}$
<center>
<h1>$\text{dim}(\ker \partial_{0})=3$</h1>
</center>

$\hspace{8mm}$
<center>
<h1>$\text{dim}(\ker \partial_{1})=1 , \text{dim}(\mathbb{Z}_2^3)=3 \Rightarrow \text{dim(V)}=\text{dim}(\mathrm {im} \,\partial_{1})+\text{dim}(\ker \partial_{1}) \because  \text{dim}(\ker \partial_{1}) =2$</h1>
</center>

$\hspace{8mm}$
<center>
<h1>$\displaystyle \text{dim}(H_{0}(\mathbb{D}))= \text{dim}(\ker \partial_{0})- \text{dim}(\mathrm {im} \,\partial_{1})= 3-2 =1 \because H_{0}(\mathbb{D})\cong \mathbb{Z}_2$</h1>
</center>

<h1><b>Bottleneck Metric</h1>
</center><br><br>

<center>
<h2>$\displaystyle W_{\infty}(\mathscr{D},\mathscr{D}')= \inf_{\gamma : \mathscr{D}\to\mathscr{D}'}  \left(\sup_{x\in \mathscr{D}} ||x-\gamma (x)||_{\infty}\right)$</h2>
</center><br><br>

<center>
<h1>Noise sensitive</h1>
</center><br><br>

```python
distance_bottleneck, x = persim.bottleneck(diagrams1[1], diagrams2[1], matching=True)
persim.visuals.bottleneck_matching(diagrams1[1], diagrams2[1], x, labels=['PTH P1', 'PTH P2'])
print('The bottleneck distance is', distance_bottleneck)
```

```python
distance_bottleneck, x = persim.bottleneck(diagrams1[1], diagrams3[1], matching=True)
persim.visuals.bottleneck_matching(diagrams1[1], diagrams3[1], x , labels=['PTH P1', 'PTH P3$'])
print('The bottleneck distance is', distance_bottleneck)
```

<center>
<h1><b>Wasserstein Metric</h1>
</center><br><br>

<center>
<h2>$\displaystyle W_{p}(\mathscr{D},\mathscr{D}')= \inf_{\gamma : \mathscr{D}\to\mathscr{D}'}  \left( \sum_{x\in \mathscr{D}} ||x-\gamma (x)||_{\infty}^p \right)^{\frac{1}{p}}$</h2>
</center><br><br>

<center>
<h1>Noise Robust</h1>
</center><br><br>

```python
distance_wasserstein, x = persim.wasserstein(diagrams1[1], diagrams2[1], matching=True)
persim.visuals.wasserstein_matching(diagrams1[1], diagrams2[1], x, labels=['PTH P1', 'PTH P2'])
print('The wasserstein distance is', distance_wasserstein)
```

```python
distance_wasserstein, x = persim.wasserstein(diagrams1[1], diagrams3[1], matching=True)
persim.visuals.wasserstein_matching(diagrams1[1], diagrams3[1], x , labels=['PTH P1', 'PTH P3$'])
print('The wasserstein distance is', distance_wasserstein)
```
